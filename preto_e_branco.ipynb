```python
import numpy as np

from matplotlib import image as mpimg
from matplotlib import pyplot as plt
from PIL import Image
```

```python
# Abra uma imagem
caminho_imagem = 'img.png'
imagem = Image.open(caminho_imagem)
```

```python
print(f"Formato da imagem: {imagem.format}")
print(f"Modo da imagem: {imagem.mode}")
print(f"Tamanho da imagem: {imagem.size}")
```

```python
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

```python
rgb = np.array(imagem)
```

```python

inGray = np.dot(rgb, [0.2989, 0.5870, 0.1140])
```

```python
plt.imshow(inGray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
inGray = inGray/255
```

```python
lim = 0.3
bin = np.where(inGray > lim, 0, 1)
```

```python
plt.imshow(bin, cmap='gray')
plt.axis('off')
plt.show()
```
